```python
from functools import reduce

'''
13. Dada una lista de Artículos con sus precio. Define las siguientes funciones:
Puedes definir más funciones si lo consideras necesario.
'''
articulos = {
    'Camisa': 20,
    'Pantalón': 30,
    'Calcetines': 5,
    'Zapatos': 50,
    'Gorra': 10,
    'Bufanda': 15,
    'Gafas': 25,
    'Reloj': 35,
    'Corbata': 40,
}

compra = ['Camisa', 'Pantalón', 'Pantalón', 'Gorra', 'Gafas', 'Corbata']


```

```python
# A. Una función que calcule el precio total de la compra

total_compra = lambda valores_compra : reduce(lambda x,y: x + y, valores_compra)
get_values_compra = [articulos.get(prenda) for prenda in compra]
suma = total_compra(get_values_compra)
print(suma)
```

```python
# B. Una función que calcule el precio total de la compra con un descuento del 10%
descuento_10 = lambda prenda: prenda*0.9
compra_descuento = list(map(descuento_10,get_values_compra))

total_compra_descuento = total_compra(compra_descuento)
print(total_compra_descuento)
```

```python
# C. Una función que calcule el precio total de la compra con un descuento del 10% si la compra supera los 100€
lista_compras = []
lista_compras.append(suma)
filtrar_compra = filter(lambda suma_compra: suma_compra > 100,lista_compras)
compra_descuento_mayor100 = list(map(descuento_10,filtrar_compra))
print(compra_descuento_mayor100)
```

```python
# D. Una función que calcule el precio total aplicando el IVA (21%)
aplicar_iva = list(map(lambda suma_tot: suma_tot *1.21, lista_compras))
print(aplicar_iva)
```
